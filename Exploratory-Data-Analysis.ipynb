# 2. Visualize the target feature (diagnosis) to see the distribution

```python
#We notice that the dataset has a column "diagnosis", which consists of "M=malignant" & "B=benign" data
#So we will split it into 2 different datasets

M = df[(df['diagnosis'] != 0)]
B = df[(df['diagnosis'] == 0)]
```

```python
y = df['diagnosis'].value_counts()
mylabels = ['Malignant', 'Benign']
plt.pie(y, labels = mylabels, autopct='%1.2f%%')
```

# 3. Visualize the distribution of some key features

```python
#Distribution of the 'mean' features

sns.displot(df, x="radius_mean")
sns.displot(df, x="texture_mean")
sns.displot(df, x="perimeter_mean")
sns.displot(df, x="area_mean")
```

```python
#Distribution of the 'se' features

sns.displot(df, x="radius_se")
sns.displot(df, x="texture_se")
sns.displot(df, x="perimeter_se")
sns.displot(df, x="area_se")
```

```python
#Distribution of the 'worst' features

sns.displot(df, x="radius_worst")
sns.displot(df, x="texture_worst")
sns.displot(df, x="perimeter_worst")
sns.displot(df, x="area_worst")
```

# 4. Correlation matrix of features

```python
# Compute the correlation matrix

corr = df.corr()


mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

The more <b>red</b> each box is, it means that the correlation coefficient is closer to 1 and the more correlated the variables are. The more <b>blue</b> each box is, it means that the correlation coefficient is closer to 0.

# 5. Positive correlated features

### We will visualize the positive correlated features ('red' color) to see their relationships

```python
#we will convert the M & B categories to 1 & 0 so we can visualize them

df.diagnosis.replace(to_replace = dict(M = 1, B = 0), inplace = True)
```

```python

sns.displot(df, x = df['perimeter_mean'], y = df['radius_worst'], hue = "diagnosis", kind='kde')
plt.title('perimeter mean vs radius worst')

sns.displot(df, x = df['area_mean'], y = df['radius_worst'], hue = "diagnosis", kind='kde')
plt.title('area mean vs radius worst')

sns.displot(df, x = df['area_worst'], y = df['radius_worst'], hue = "diagnosis", kind='kde')
plt.title('area worst vs radius worst')

sns.displot(df, x = df['texture_mean'], y = df['texture_worst'], hue = "diagnosis", kind='kde')
plt.title('texture mean vs texture worst')
```

We can see from these distribution visualizations, that positive correlations can be concluded through linear relationships

# 6. Uncorrelated features

### We will do the same process for the uncorrelated features to see their relationships

```python
sns.displot(df, x = df['smoothness_mean'], y = df['texture_mean'], hue = "diagnosis", kind='kde')
plt.title('smoothness mean vs texture mean')

sns.displot(df, x = df['texture_mean'], y = df['symmetry_mean'], hue = "diagnosis", kind='kde')
plt.title('texture mean vs symmetry mean')

sns.displot(df, x = df['texture_mean'], y = df['symmetry_se'], hue = "diagnosis", kind='kde')
plt.title('texture mean vs symmetry se')

sns.displot(df, x = df['radius_mean'], y = df['fractal_dimension_worst'], hue = "diagnosis", kind='kde')
plt.title('radius mean vs fractal dimension_worst')
```

As we see, if two variables are not correlated there is no linear relationship between them

# 7. Negative correlated features

### We will do the same process for the negative correlated ('blue' color) features to see their relationships

```python
sns.displot(df, x = df['area_mean'], y = df['fractal_dimension_mean'], hue = "diagnosis", kind='kde')
plt.title('area mean vs fractal dimension mean')

sns.displot(df, x = df['radius_mean'], y = df['fractal_dimension_mean'], hue = "diagnosis", kind='kde')
plt.title('radius mean vs fractal dimension mean')

sns.displot(df, x = df['area_mean'], y = df['smoothness_se'], hue = "diagnosis", kind='kde')
plt.title('area mean vs smoothness se')

sns.displot(df, x = df['smoothness_se'], y = df['perimeter_mean'], hue = "diagnosis", kind='kde')
plt.title('smoothness se vs perimeter mean')
```

When there is a negative correlation between two variables it means that when one variable increases, the other one decreases.
